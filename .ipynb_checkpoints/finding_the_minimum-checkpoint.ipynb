```python
import numpy as np
import matplotlib.pyplot as plt
```

Let F(x) be as follows:$$ F(x, y) = (x^2 + y − 11)^2 + (x + y^2 − 7)^2 $$

Find a minimum of the previous function starting from point [4,-4] using the
following methods in the minimize function (of scipy.optimize module):
<ol>
    <li> Powell's method </li>
    <li>  Conjugate Gradient </li>
    <li>  BFGS</li>
</ol>


## using powell's method

```python
from scipy.optimize import minimize
%matplotlib inline

def Q(v):
    x,y = v[0],v[1]
    
    return x**4 + y**4 +2*x**2*y -21*x**2 -13*y**2-22*y+2*x*y**2-14*x+170

x0 = np.array([4, -4])
res = minimize(Q, x0, method='Powell', options={'disp':True, 'return_all':True})
print("The minima using Powell's: ",res.x)
xs = res.allvecs


LB = -5
UB = 5
x = np.linspace(LB, UB, 50)
y = np.linspace(LB, UB, 50)
X, Y = np.meshgrid(x, y)

Z = Q([X, Y])

fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(111)
ax1.contour(X, Y, Z, levels=20)
ax1.scatter([x[0] for x in xs], [x[1] for x in xs], c=list(range(len(res.allvecs))), marker='x')
ax1.plot([x[0] for x in xs], [x[1] for x in xs])
ax1.set_title("The minima using Powell's: "+str(res.x))
plt.show()
```

## using Conjugate Gradient

```python
def Q(v):
    x,y = v[0],v[1]
    
    return x**4 + y**4 +2*x**2*y -21*x**2 -13*y**2-22*y+2*x*y**2-14*x+170

def dxQ(v):
    x,y = v[0],v[1]
    return 4*x**3 + 4*x*y - 42*x + 2*y**2 - 14

def dyQ(v):
    x,y = v[0],v[1]
    return 4*y**3 + 2*x**2 -26*y -22 + 4*x*y
    
def dQ(v):
    d = np.array([dxQ(v),dyQ(v)])
    return d

x0 = np.array([4, -4])
res = minimize(Q, x0,jac=dQ, method='CG', options={'disp':True, 'return_all':True})
print("The minima using CG: ",res.x)
xs = res.allvecs


LB = -5
UB = 5
x = np.linspace(LB, UB, 50)
y = np.linspace(LB, UB, 50)
X, Y = np.meshgrid(x, y)

Z = Q([X, Y])

fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(111)
ax1.contour(X, Y, Z, levels=20)
ax1.scatter([x[0] for x in xs], [x[1] for x in xs], c=list(range(len(res.allvecs))), marker='x')
ax1.plot([x[0] for x in xs], [x[1] for x in xs])
ax1.set_title("The minima using Conjugate Gradient: "+str(res.x))
plt.show()
```

## using BFGS

```python
def Q(v):
    x,y = v[0],v[1]
    
    return x**4 + y**4 +2*x**2*y -21*x**2 -13*y**2-22*y+2*x*y**2-14*x+170

def dxQ(v):
    x,y = v[0],v[1]
    return 4*x**3 + 4*x*y - 42*x + 2*y**2 - 14

def dyQ(v):
    x,y = v[0],v[1]
    return 4*y**3 + 2*x**2 -26*y -22 + 4*x*y
    
def dQ(v):
    d = np.array([dxQ(v),dyQ(v)])
    return d

x0 = np.array([4, -4])
res = minimize(Q, x0,jac=dQ, method='BFGS', options={'disp':True, 'return_all':True})
print("The minima using BFGS: ",res.x)
xs = res.allvecs


LB = -5
UB = 5
x = np.linspace(LB, UB, 50)
y = np.linspace(LB, UB, 50)
X, Y = np.meshgrid(x, y)

Z = Q([X, Y])

fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(111)
ax1.contour(X, Y, Z, levels=20)
ax1.scatter([x[0] for x in xs], [x[1] for x in xs], c=list(range(len(res.allvecs))), marker='x')
ax1.plot([x[0] for x in xs], [x[1] for x in xs])
ax1.set_title("The minima using BFGS: "+str(res.x))
plt.show()
```
