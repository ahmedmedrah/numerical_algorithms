# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 3 + 1 * x  + np.random.normal(0, 1, size=100)

plt.figure(figsize=(10, 8))
plt.scatter(x, y, c='b', marker='x')
plt.show()
```

## Fit a line

Use the Moore-Penrose inverse to find the best line that fits the data.

```python
x = x.reshape(-1,1)
A = np.ones_like(x)
A =np.append(A, x, axis=1)
# np.hstack((A,x)).shape
A.shape
```

```python
A_inv = np.linalg.pinv(A)
```

```python
coefs = A_inv.dot(y)
coefs
```

```python
yhat = coefs[0] + coefs[1]*x

plt.plot(x, y, '*')
plt.plot(x, yhat)

plt.show()
```
